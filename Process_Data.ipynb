```python
def get_data_at_index(csv_line, index):
    arr = csv_line.split(",")
    return arr[index]

# Creates array of ids found in first column of "filename" which is a .csv file
# matches_array is an array of ids containing ids shared across data, [] if first set of data
# orig is a boolean and is True if it is the first id array being made and False otherwise
# if orig is true, returns an array of all ids in data
# if orig is false, returns array of all ids of new data that is also in orig_array
def create_data_array_from_file(filename, matches_array, orig, index):
    fp = open(filename, "r")
    ids = []
    count = 1
    while (True):
        line = fp.readline()
        if (not line):      # Break while loop if end of line is reached.
            break
        if (count > 1):     # Skip the first line with just information 
            if (orig):      
                ids.append(get_data_at_index(line, index))  # if first set of data append all ids
            else:
                check = get_data_at_index(line, index)      
                if (check in matches_array): # if not first set only append data already in matches_array
                    ids.append(check)
        count += 1
    return ids
```

```python
import copy

# Code taken from internet
# https://www.geeksforgeeks.org/print-all-possible-combinations-of-r-elements-in-a-given-array-of-size-n/
# Modified to put all combinations in the new input called "output"

# The main function that prints  
# all combinations of size r in  
# arr[] of size n. This function  
# mainly uses combinationUtil() 
def getCombinations(arr, n, r, output): 
      
    # A temporary array to  
    # store all combination 
    # one by one 
    data = [0]*r; 
  
    # Print all combination  
    # using temprary array 'data[]' 
    combinationUtil(arr, data, 0,  
                    n - 1, 0, r, output); 
  
# arr[] ---> Input Array 
# data[] ---> Temporary array to 
#         store current combination 
# start & end ---> Staring and Ending 
#             indexes in arr[] 
# index ---> Current index in data[] 
# r ---> Size of a combination  
# to be printed  
def combinationUtil(arr, data, start,  
                    end, index, r, output): 
                          
    # Current combination is ready  
    # to be printed, print it 
    if (index == r): 
        output.append(copy.copy(data)); 
        return; 
  
    # replace index with all 
    # possible elements. The 
    # condition "end-i+1 >=  
    # r-index" makes sure that  
    # including one element at 
    # index will make a combination  
    # with remaining elements at  
    # remaining positions 
    i = start;  
    while(i <= end and end - i + 1 >= r - index): 
        data[index] = arr[i]; 
        combinationUtil(arr, data, i + 1,  
                        end, index + 1, r, output); 
        i += 1; 
```

```python
# Get all the possible combinations of the data containing cp ids
combos = []
for i in range(1,7):
    getCombinations(["Data 1.csv","Data 2.csv","Data 3.csv","Data 4.csv", "Data 8.csv", "Data 10.csv"], 6, i, combos)
for i in combos:
    print(i)
```

```python
# Find cg islands common to all data in each combination
fp = open("ids.csv", "w")
for combination in combos:
    if (len(combination) > 1): # Only look at combinations with multiple data sets
        matches = []
        for i in range(len(combination)):
            if (i == 0):
                matches = create_data_array_from_file(combination[i], matches, True, 0)
            else:
                matches = create_data_array_from_file(combination[i], matches, False, 0)
        if (len(matches) > 0):
            fp.write(" - ".join(combination) + ":\n")
            fp.write(",".join(matches) + "\n\n")
fp.close()
    
```

```python
# Get all the possible combinations of the data containing gene names
combos2 = []
for i in range(1,10):
    getCombinations(["Data 2.csv","Data 3.csv","Data 4.csv","Data 5.csv","Data 6.csv", "Data 7.csv", "Data 8.csv", "Data 9.csv", "Data 10.csv"], 9, i, combos2)
for i in combos2:
    print(i)
```

```python
# Find gene names common to all data in each combination
fp = open("names.csv", "w")
for combination in combos2:
    if (len(combination) > 1): # Only look at combinations with multiple data sets
        matches = []
        orig = False
        for i in range(len(combination)):
            file = combination[i]
            if (i == 0):
                orig = True
            if (file == "Data 2.csv" or file == "Data 3.csv" or file == "Data 4.csv"):
                matches = create_data_array_from_file(file, matches, orig, 6)
            elif (file == "Data 5.csv" or file == "Data 6.csv" or file == "Data 7.csv" or file == "Data 10.csv"):
                matches = create_data_array_from_file(file, matches, orig, 1)
            elif (file == "Data 8.csv"):
                matches = create_data_array_from_file(file, matches, orig, 4)
            elif (file == "Data 9.csv"):
                matches = create_data_array_from_file(file, matches, orig, 0)
        matches = set(matches)
        if (len(matches) > 0):
            fp.write(" - ".join(combination) + ":\n")
            fp.write(",".join(matches) + "\n\n")
fp.close()
```
